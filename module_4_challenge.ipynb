## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

# Created a new column 'Line Subtotal' by multiplying 'Unit Price' and 'Quantity.'
df['Line Subtotal'] = df['Unit Price'] * df['Quantity']

# Selected relevant columns for display.
select_columns = ['Unit Price', 'Quantity', 'Line Subtotal']

# Displayed the DataFrame with the selected columns.
df[select_columns].head()
```

```python
# I created a new column, 'Total Weight' that references: 'Unit Weight' and 'Quantity'
df['Total Weight'] = df['Unit Weight'] * df['Quantity']

# Displayed the first few rows to ensure 'Total Weight' was created properly.   
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# Defined a function to calculate the shipping price based on 'Total Weight.'
def calculate_shipping_price(total_weight):
    if total_weight > 50:
        return 7 * total_weight
    else:
        return 10 * total_weight

# Applied the shipping price calculation function to the 'Total Weight' column.
df['Shipping Price'] = df['Total Weight'].apply(calculate_shipping_price)

# Selected relevant columns for display.
select_columns_b = ['Unit Price', 'Unit Weight', 'Quantity', 'Total Weight', 'Shipping Price']

# Displayed the first few rows with the selected columns.    
df[select_columns_b].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# Created a column 'Total Price' for each line adding 'Line Subtotal'to 
# 'Shipping Price', and added a sales tax of 9.25%.
df['Total Price'] = ((df['Line Subtotal'] + df['Shipping Price']) * 1.0925).round(2)

# Selected relevant columns for display.
select_columns_c = ['Line Subtotal', 'Shipping Price', 'Total Price']

# Displayed the first few rows with the selected columns.
df[select_columns_c].head()
```

```python
# Create a column for the cost 
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

# Created a column 'Total Cost' for each line multiplying 'Unit Cost' with 'Quantity', 
# and adding 'Shipping Price." Then rounded output data to 2 decimals.
df['Total Cost'] = (df['Unit Cost'] * df['Quantity'] + df['Shipping Price']).round(2)

# Selected relevant columns for display.
select_columns_d = ['Unit Cost', 'Quantity', 'Shipping Price', 'Total Price', 'Total Cost']

# Displayed the first few rows with the selected columns.
df[select_columns_d].head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

# Created a column, 'Gross Profit' for each line using 'Total Price' to subtract 'Total Cost.'
df['Gross Profit'] = (df['Total Price'] - df['Total Cost']).round(2)

# Selected relevant columns for display.
select_columns_e = ['Total Price', 'Total Cost', 'Gross Profit']

# Displayed the first 5 rows of the selected columns.
df[select_columns_e].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? 
# Check your work from Part 1 for client ids.

# Referenced 'client_id_count' from Part 1.
top_5 = client_id_count.head(5).index

# Filtered the DataFrame for the top 5 clients.
top_5_df = df[df['Client ID'].isin(top_5)]

# Calculated the total spending for the top 5 clients.
top_5_total_spend = top_5_df.groupby('Client ID')['Total Price'].sum()

# Formated the total spending as a dollar amount.
usd_total_spend = top_5_total_spend.map('${:,.2f}'.format)

# Displayed the total spending for the top 5 clients.
print(usd_total_spend)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Created a DataFrame with totals for the top 5 clients.
sum_a_df = pd.DataFrame({
    'Total Quantity Purchased': top_5_df.groupby('Client ID')['Quantity'].sum(),
    'Total Shipping Price': top_5_df.groupby('Client ID')['Shipping Price'].sum(),
    'Total Price': top_5_df.groupby('Client ID')['Total Price'].sum(),
    'Total Profit': top_5_df.groupby('Client ID')['Gross Profit'].sum()
}).round(2)

# Displayed the DataFrame with totals for the top 5 clients.
sum_a_df.head()
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Previous column names are adequately formatted.
# Formatted previous DataFrame to display in millions $USD to 3 decimal places.
mm_sum_a_df = pd.DataFrame({
    'Total Quantity Purchased': top_5_df.groupby('Client ID')['Quantity'].sum(),
    'Total Shipping Price (MM)': (top_5_df.groupby('Client ID')['Shipping Price'].sum() / 1_000_000).map('${:,.3f}'.format),
    'Total Price (MM)': (top_5_df.groupby('Client ID')['Total Price'].sum() / 1_000_000).round(2).map('${:,.3f}'.format),
    'Total Profit (MM)': (top_5_df.groupby('Client ID')['Gross Profit'].sum() / 1_000_000).round(2).map('${:,.3f}'.format)
})

# Displayed the updated DataFrame
mm_sum_a_df.head()
```

```python
# Sort the updated data by "Total Profit" form highest to lowest.

# Ammended above data in descending order.
sort_mm_sum_a_df = mm_sum_a_df.sort_values(by='Total Profit (MM)', ascending=False)

# Displayed the updated DataFrame
sort_mm_sum_a_df.head()
```
